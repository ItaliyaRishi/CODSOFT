```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd. read_csv("advertising.csv")
df. head()
```

```python
df.describe
```

```python
df.shape
```

```python
#visualisations
sns.pairplot(df, x_vars=['TV', 'Radio', 'Newspaper'], y_vars='Sales', kind='scatter', palette='husl')
plt.show()
#observation - As spending more on TV advertisements correlates with higher sales, the relationship is more consistent. However, for newspaper and radio ads, the impact on sales is less predictable.
```

```python
sns.heatmap(df.corr(),annot = True)
plt. show()
# TV is strongly correlated with Sales 0.9
```

```python
df['Radio'].plot.hist(bins=10, color="skyblue")
plt.xlabel("Radio")
plt.ylabel("Frequency")
plt.title("Histogram of Radio Advertisement Budget")
plt.show()
```

```python
df['Newspaper'].plot.hist(bins=10, color="lightgreen")
plt.xlabel("Newspaper")
plt.ylabel("Frequency")
plt.title("Histogram of Newspaper Advertisement Budget")
plt.show()
# lower the advertising cost on newspaper, higher the sales.
```
